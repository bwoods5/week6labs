This notebook contains code for a magic crystal ball that has answers for any question you may have. Beware, luck may not be on your side!



## Import Modules

```python
import random as rand
```

## Create Variables


```python
fortune_responses = [
    # Standard responses
    "It is certain",
    "It is decidedly so",
    "That question requires greater monetary compensation",
    "Reply is hazy; try again, with more money",
    "Concentrate and ask again",
    "My reply is no",
    "Outlook is not so good",
    "Very doubtful",
    # My custom responses
    "Why even ask? That's a resounding yes!",
    "... I don't think so",
    "I couldn't hear you, could you repeat that?"
]

question_asked = False  # keep track if a question has been asked yet or not
service_cost = 0.0       # keep track of costs up to this point

questions = []  # the questions asked
responses = []  # the responses provided 
```


## Create Nested Loop

```python

while True:
    menu_text = "Ask another" if question_asked else "Ask a"
    print("\n1.", menu_text, "question")
    print("2. Review questions and responses")
    print("3. Complete services and pay\n")

    menu_choice = input("Select an option (1, 2, or 3): ")

    if menu_choice == '1':
        print("\nAsk what you seek to know...")
        question = input()
        questions.append(question)
        question_asked = True

        response = rand.choice(fortune_responses)
        responses.append(response)
        print(response + "\n")

        service_cost += 30.00

    elif menu_choice == '2':
        if not questions:
            print("\nYou have not asked any questions yet.\n")
        else:
            print()
            for i in range(len(questions)):
                print(f"Question: \"{questions[i]}\"\nResponse: \"{responses[i]}\"\n")

    elif menu_choice == '3':
        print("\nThank you for visiting the all-seeing computer.\n")
        print(f"You asked {len(questions)} questions. You owe ${service_cost:.2f} for today's services.\n")
        break

    else:
        print("\nInvalid option. Please choose 1, 2, or 3.\n")
```
