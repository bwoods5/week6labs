This notebook produces a user menu that stores, queries, and adds student information. It will allow the user to view existing student information, query out GPA details, add in new students along with their information, and add in student violations. It is a one stop shop for learning management.

## Initialize student records


```python
students = [                                   #Assigning students and their information
    {"first_name": "Carmen", "last_name": "Berzatto", "major": "Marketing", "gpa": 3.91, "violation": None},
    {"first_name": "Sydney", "last_name": "Adamu", "major": "Business", "gpa": 3.72, "violation": None},
    {"first_name": "Neil", "last_name": "Fak", "major": "Communications", "gpa": 2.95, "violation": None},
    {"first_name": "Marcus", "last_name": "Brooks", "major": "Economics", "gpa": 3.23, "violation": None},
]


violations_list = [                              #Assigning potential violations
    "Distributing Course Materials",
    "Excessive Plagiarism",
    "Excessive Absences",
    "Using Technology During Exam",
    "Collabterating on Exam"
]
```

## Menu loop


```python
while True:                                     #While loop to keep the program running
    print("\n Student Management System")
    print("1. Student Information Query")
    print("2. Student Information Entry")
    print("3. Query GPA Metrics")
    print("4. Academic Violations")
    print("5. Quit Program")
    choice = input("\nSelect an option (1-5): ")

    if choice == '1':                           #Choice for querying student information
        print("\n All Student Information")
        for student in students:
            print(f"{student['first_name']} {student['last_name']} | Major: {student['major']} | GPA: {student['gpa']} | Violation: {student['violation']}")

    elif choice == '2':                         #Choice for entering new student information            
        print("\n Add New Student")
        first = input("First name: ")
        last = input("Last name: ")
        major = input("Major: ")
        gpa = float(input("GPA: "))
        students.append({"first_name": first, "last_name": last, "major": major, "gpa": gpa, "violation": None})
        print("Student added successfully!")

    elif choice == '3':                         #Choice for querying GPA metrics
        gpas = [s['gpa'] for s in students]
        print("\n--- GPA Metrics ---")
        print(f"Max GPA: {max(gpas)}")
        print(f"Min GPA: {min(gpas)}")
        print(f"Average GPA: {sum(gpas)/len(gpas):.2f}")

    elif choice == '4':                        #Choice for assigning academic violations
        print("\n Assign Academic Violations")
        for i, student in enumerate(students):
            print(f"{i+1}. {student['first_name']} {student['last_name']}")
        selected = int(input("Select a student by number: ")) - 1
        print("Select a violation:")
        for i, v in enumerate(violations_list):
            print(f"{i+1}. {v}")
        v_choice = int(input("Violation number: ")) - 1
        students[selected]['violation'] = violations_list[v_choice]
        print("Violation assigned successfully.")

    elif choice == '5':                       #Choice for quitting the program
        print("\nExiting program. Goodbye!")
        break

    else:
        print("\nInvalid selection. Try again.")
```
